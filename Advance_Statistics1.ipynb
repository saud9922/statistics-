````python
"""
### ##Theoretical Questions##

##1. [cite\_start]What is a random variable in probability theory?## [cite: 6]
A random variable is a variable whose value is a numerical outcome of a random phenomenon. It assigns a numerical value to each possible outcome in a sample space. For instance, in a coin toss experiment, a random variable could be defined to be 1 if the outcome is "heads" and 0 if the outcome is "tails."

##2. [cite\_start]What are the types of random variables?## [cite: 7]
Random variables can be broadly classified into two main types:

  * ##Discrete Random Variable:## A variable that can take on a finite or countably infinite number of distinct values. Examples include the number of heads in a series of coin flips or the number of defective items in a batch.
  * ##Continuous Random Variable:## A variable that can take on any value within a given range or interval. Examples include the height of a person, the temperature of a room, or the time it takes to complete a task.

##3. [cite\_start]What is the difference between discrete and continuous distributions?## [cite: 8]
The primary difference lies in the nature of the random variable they describe:

  * ##Discrete Distribution:## Describes the probability of occurrence for each possible value of a discrete random variable. The probabilities are represented by a probability mass function (PMF), and the sum of all probabilities is equal to 1.
  * ##Continuous Distribution:## Describes the probabilities for the possible values of a continuous random variable. The probability is represented by a probability density function (PDF), where the probability of the variable falling within a specific range is the area under the curve of the PDF over that range. The total area under the curve is 1.

##4. [cite\_start]What are probability distribution functions (PDF)?## [cite: 9]
A Probability Density Function (PDF) is a function associated with a continuous random variable. The primary purpose of a PDF is to describe the relative likelihood that the random variable will take on a particular value. The probability of the variable falling within a specific interval is given by the integral (the area under the curve) of the PDF over that interval. The value of the PDF at any single point is not the probability of that point, but rather its density; the probability of any single point for a continuous variable is zero.

##5. [cite\_start]How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?## [cite: 10]
A Cumulative Distribution Function (CDF) and a Probability Density Function (PDF) provide different types of information about a probability distribution:

  * ##PDF (Probability Density Function):## As mentioned, the PDF of a continuous random variable indicates the relative likelihood of a random variable taking a specific value. For a discrete variable, the equivalent is the Probability Mass Function (PMF), which gives the probability of a specific outcome.
  * ##CDF (Cumulative Distribution Function):## The CDF gives the probability that a random variable $X$ will take a value less than or equal to a specific value $x$. It is defined as $F(x) = P(X \\le x)$. The CDF accumulates the probabilities up to the value $x$. For a continuous variable, the CDF is the integral of the PDF from negative infinity to $x$.

##6. [cite\_start]What is a discrete uniform distribution?## [cite: 11]
A discrete uniform distribution is a probability distribution where a finite number of values are equally likely to be observed. If a random variable can take any of $n$ distinct values, then the probability of observing each value is $1/n$. A classic example is the roll of a fair six-sided die, where each outcome (1, 2, 3, 4, 5, or 6) has a probability of $1/6$.

##7. [cite\_start]What are the key properties of a Bernoulli distribution?## [cite: 12]
A Bernoulli distribution is a discrete probability distribution for a random variable which takes the value 1 with probability $p$ and the value 0 with probability $1-p$. Its key properties are:

  * ##Two Outcomes:## There are only two possible outcomes for each trial, often termed "success" (value 1) and "failure" (value 0).
  * ##Single Trial:## It models the outcome of a single experiment or trial.
  * ##Probability Parameter:## The distribution is defined by a single parameter, $p$, which is the probability of success.
  * ##Expected Value:## The expected value (mean) is $E[X] = p$.
  * ##Variance:## The variance is $Var(X) = p(1-p)$.

##8. [cite\_start]What is the binomial distribution, and how is it used in probability?## [cite: 13]
The binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials. It is used in probability to calculate the likelihood of observing a specific number of successful outcomes when the experiment is repeated multiple times. The distribution is defined by two parameters:

  * $n$: The total number of trials.
  * $p$: The probability of success in a single trial.

The probability of getting exactly $k$ successes in $n$ trials is given by the formula:
$$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$

##9. [cite\_start]What is the Poisson distribution and where is it applied?## [cite: 14]
The Poisson distribution is a discrete probability distribution that expresses the probability of a given number of events occurring in a fixed interval of time or space if these events occur with a known constant mean rate and independently of the time since the last event. It is characterized by a single parameter, $\\lambda$ (lambda), which is the mean number of events in the interval.
##Applications:##

  * The number of phone calls received by a call center per hour.
  * The number of emails received per day.
  * The number of defects in a manufactured product.
  * The number of accidents at an intersection in a week.

##10. [cite\_start]What is a continuous uniform distribution?## [cite: 15]
A continuous uniform distribution is a probability distribution where all outcomes within a certain range are equally likely. It is defined by two parameters, $a$ and $b$, which are the minimum and maximum values (the endpoints of the interval). The probability density function is constant over the interval $[a, b]$ and zero everywhere else.

##11. [cite\_start]What are the characteristics of a normal distribution?## [cite: 16]
The normal distribution, also known as the Gaussian distribution, is a continuous probability distribution with the following key characteristics:

  * ##Bell-Shaped Curve:## The graph of its probability density function is a symmetric, bell-shaped curve.
  * ##Symmetry:## It is symmetric about its mean. The mean, median, and mode are all equal and located at the center of the distribution.
  * ##Parameters:## It is defined by two parameters: the mean ($\\mu$) and the standard deviation ($\\sigma$). The mean determines the center of the distribution, and the standard deviation determines the spread.
  * ##Empirical Rule:## Approximately 68% of the data falls within one standard deviation of the mean, 95% within two standard deviations, and 99.7% within three standard deviations.
  * ##Asymptotic:## The tails of the curve extend indefinitely and never touch the horizontal axis.

##12. [cite\_start]What is the standard normal distribution, and why is it important?## [cite: 17]
The standard normal distribution is a special case of the normal distribution where the mean ($\\mu$) is 0 and the standard deviation ($\\sigma$) is 1.
It is important for several reasons:

  * ##Standardization:## Any normal distribution can be converted to a standard normal distribution by transforming the values using the Z-score formula: $Z = (X - \\mu) / \\sigma$.
  * ##Simplification:## This transformation allows for the comparison of scores from different normal distributions.
  * ##Probability Calculation:## It simplifies the process of finding probabilities. Probabilities for any normal distribution can be found using the standard normal distribution table (Z-table) or statistical software.

##13. [cite\_start]What is the Central Limit Theorem (CLT), and why is it critical in statistics?## [cite: 18]
The Central Limit Theorem (CLT) is a fundamental theorem in statistics which states that the distribution of sample means of a large number of samples taken from a population will be approximately normally distributed, regardless of the shape of the population's distribution, as long as the sample size is sufficiently large (usually $n \\ge 30$).
It is critical because:

  * ##Foundation for Inference:## It allows us to make inferences about a population mean using the sample mean, even if we don't know the shape of the population's distribution.
  * ##Hypothesis Testing:## It is the basis for many hypothesis tests and the construction of confidence intervals.
  * ##Real-World Application:## Many statistical procedures rely on the assumption of normality, and the CLT allows these procedures to be used for a wide variety of problems.

##14. [cite\_start]How does the Central Limit Theorem relate to the normal distribution?## [cite: 19]
The Central Limit Theorem establishes a connection between any population distribution and the normal distribution. It asserts that the sampling distribution of the sample mean (the distribution of means from many samples) will tend to be a normal distribution as the sample size increases. This holds true even if the original population distribution is not normal. This relationship allows statisticians to use the properties of the normal distribution to analyze sample data and make inferences about the population.

##15. [cite\_start]What is the application of Z statistics in hypothesis testing?## [cite: 20]
Z-statistics are used in hypothesis testing to determine whether there is a statistically significant difference between a sample mean and a known or hypothesized population mean. This is typically done in a Z-test when the population standard deviation is known and the sample size is large. The Z-statistic measures how many standard deviations the sample mean is away from the population mean. This value is then compared to a critical value from the standard normal distribution to decide whether to reject or fail to reject the null hypothesis.

##16. [cite\_start]How do you calculate a Z-score, and what does it represent?## [cite: 21]
A Z-score is calculated using the formula:
$$Z = \frac{(X - \mu)}{\sigma}$$
where:

  * $X$ is the value of an individual data point.
  * $\\mu$ is the mean of the population.
  * $\\sigma$ is the standard deviation of the population.

A Z-score represents the number of standard deviations a data point is from the mean of its distribution.

  * A positive Z-score indicates the data point is above the mean.
  * A negative Z-score indicates the data point is below the mean.
  * A Z-score of 0 indicates the data point is equal to the mean.

##17. [cite\_start]What are point estimates and interval estimates in statistics?## [cite: 22]

  * ##Point Estimate:## A point estimate is a single value (a statistic) that is used to estimate an unknown population parameter. For example, the sample mean ($\\bar{x}$) is a point estimate of the population mean ($\\mu$).
  * ##Interval Estimate:## An interval estimate provides a range of values within which the population parameter is expected to lie, along with a certain level of confidence. This range is called a confidence interval. For example, a 95% confidence interval for the population mean gives a range of values that we are 95% confident contains the true population mean.

##18. [cite\_start]What is the significance of confidence intervals in statistical analysis?## [cite: 23]
Confidence intervals are significant because they:

  * ##Quantify Uncertainty:## They provide a measure of the uncertainty associated with a point estimate. A wider interval implies more uncertainty, while a narrower interval suggests more precision.
  * ##Provide a Range of Plausible Values:## Instead of a single number, they give a range of plausible values for the unknown population parameter.
  * ##Aid in Decision Making:## They are used in hypothesis testing. If a hypothesized value for a parameter falls outside the confidence interval, it provides evidence against the hypothesis.

##19. [cite\_start]What is the relationship between a Z-score and a confidence interval?## [cite: 24]
Z-scores are a key component in calculating confidence intervals for a population mean when the population standard deviation is known or the sample size is large. The formula for a confidence interval for the mean is:
$$\text{Confidence Interval} = \bar{x} \pm Z \times \frac{\sigma}{\sqrt{n}}$$
Here, the Z-score (also called the critical value) corresponds to the desired level of confidence. For example, for a 95% confidence interval, the Z-score is approximately 1.96. This Z-score defines the boundaries of the interval in terms of standard errors from the sample mean.

##20. [cite\_start]How are Z-scores used to compare different distributions?## [cite: 25]
Z-scores standardize data points from different normal distributions, allowing for a meaningful comparison. By converting values from different datasets (with different means and standard deviations) into Z-scores, we can place them on a common scaleâ€”the standard normal distribution. This allows us to compare relative positions. For example, we can determine whether a student's score of 85 on a test with a mean of 75 and a standard deviation of 5 is better or worse than another student's score of 90 on a different test with a mean of 80 and a standard deviation of 10.

##21. [cite\_start]What are the assumptions for applying the Central Limit Theorem?## [cite: 26]
The key assumptions for applying the Central Limit Theorem are:

  * ##Random Sampling:## The samples must be drawn randomly from the population.
  * ##Independence:## The observations within each sample must be independent of each other. When sampling without replacement, the sample size should be less than 10% of the population to ensure independence.
  * ##Sample Size:## The sample size must be sufficiently large. A common rule of thumb is a sample size of 30 or more ($n \\ge 30$). However, if the population distribution is already close to normal, a smaller sample size may suffice.

##22. [cite\_start]What is the concept of expected value in a probability distribution?## [cite: 27]
The expected value of a random variable is the long-run average value of repetitions of the experiment it represents. It is a weighted average of all possible values that the random variable can take on, where the weights are the probabilities of each value. For a discrete random variable $X$ with possible values $x\_i$ and corresponding probabilities $p\_i$, the expected value $E[X]$ is calculated as:
$$E[X] = \sum_i x_i p_i$$
The expected value is also referred to as the mean of the probability distribution.

##23. [cite\_start]How does a probability distribution relate to the expected outcome of a random variable?## [cite: 28]
A probability distribution specifies all possible outcomes of a random variable along with their corresponding probabilities. The expected outcome (or expected value) is a summary measure of this distribution. It represents the central tendency or the average value we would expect to see if we performed the random experiment many times. The probability distribution provides the full picture of the variability and likelihood of all outcomes, while the expected value condenses this information into a single number representing the long-term average outcome.

-----
"""
````

````python
##9. [cite\_start]Write a Python function to calculate Z-scores from a dataset and plot them.## [cite: 38]


import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a dataset and plots their distribution.

    Args:
        data (array-like): The input dataset.
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    print(f"Mean of original data: {mean:.2f}")
    print(f"Standard Deviation of original data: {std:.2f}")
    print(f"Mean of Z-scores: {np.mean(z_scores):.2f}")
    print(f"Standard Deviation of Z-scores: {np.std(z_scores):.2f}")

    # Plotting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.hist(data, bins=30, alpha=0.7, color='blue')
    ax1.set_title('Original Data Distribution')
    ax1.set_xlabel('Value')
    ax1.set_ylabel('Frequency')

    ax2.hist(z_scores, bins=30, alpha=0.7, color='green')
    ax2.set_title('Z-Score Distribution')
    ax2.set_xlabel('Z-Score')

    plt.tight_layout()
    plt.show()
    return z_scores

# Example usage with some sample data
sample_data = np.random.normal(50, 10, 1000)
z_scores = calculate_and_plot_z_scores(sample_data)
````

````python
#### **11. Bayesian Inference with Bayes' Theorem**


def bayesian_inference(prior, likelihood, evidence):
    """
    Simple implementation of Bayes' Theorem.
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Medical Diagnosis
# P(A) = Prior probability of having the disease
# P(B|A) = Probability of testing positive given the disease (sensitivity)
# P(B) = Overall probability of testing positive

p_disease = 0.01
p_positive_given_disease = 0.99 # Sensitivity
p_positive_given_no_disease = 0.05 # False positive rate
p_no_disease = 1 - p_disease

# P(B) = P(B|A)P(A) + P(B|not A)P(not A)
p_positive = (p_positive_given_disease * p_disease) + (p_positive_given_no_disease * p_no_disease)

# Calculate P(Disease | Positive Test)
p_disease_given_positive = bayesian_inference(p_disease, p_positive_given_disease, p_positive)

print(f"Posterior probability of having the disease given a positive test: {p_disease_given_positive:.4f}")
```

-----
````

````python
#### **12. Chi-square Test for Independence**


import numpy as np
from scipy.stats import chi2_contingency

# Observed data in a contingency table
# Rows: Smoker, Non-smoker; Columns: Lung Cancer, No Lung Cancer
observed = np.array([[50, 10], [20, 120]])

chi2, p, dof, expected = chi2_contingency(observed)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)

if p < 0.05:
    print("\nReject the null hypothesis: There is a significant association between smoking and lung cancer.")
else:
    print("\nFail to reject the null hypothesis.")

````

````python
#### **13. Calculate Expected Frequencies for Chi-square Test**


import numpy as np

def expected_frequencies(observed_table):
    """
    Calculates expected frequencies for a Chi-square test.
    """
    row_totals = observed_table.sum(axis=1)
    col_totals = observed_table.sum(axis=0)
    grand_total = observed_table.sum()
    
    expected_table = np.outer(row_totals, col_totals) / grand_total
    return expected_table

# Example usage
observed = np.array([[30, 10], [15, 25]])
expected = expected_frequencies(observed)
print("Observed Frequencies:\n", observed)
print("\nExpected Frequencies:\n", expected)


````

````python
#### **14. Goodness-of-Fit Test**


from scipy.stats import chisquare

# Observed frequencies of dice rolls
observed = [9, 11, 8, 12, 10, 10] # Total 60 rolls
# Expected frequencies for a fair die
expected = [10, 10, 10, 10, 10, 10]

chi2, p = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Reject the null hypothesis: The die is likely not fair.")
else:
    print("Fail to reject the null hypothesis: The die appears to be fair.")

````

````python
##15. [cite\_start]Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.## [cite: 41]
*(Note: The CLT describes the sampling distribution of the mean from *any* population, so verifying it with a normal population shows that the sampling distribution of the mean is also normal, with a smaller standard deviation.)*


import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Population parameters (already normal)
pop_mean = 10
pop_std = 5
population = np.random.normal(pop_mean, pop_std, size=100000)

# Sampling parameters
sample_size = 30
num_samples = 2000
sample_means = [np.mean(np.random.choice(population, sample_size)) for _ in range(num_samples)]

# Theoretical values from CLT
clt_mean = pop_mean
clt_std = pop_std / np.sqrt(sample_size)

# Print comparison
print(f"Population Mean: {pop_mean}, Population Std Dev: {pop_std}")
print(f"Mean of Sample Means: {np.mean(sample_means):.4f}, Std Dev of Sample Means: {np.std(sample_means):.4f}")
print(f"CLT Theoretical Mean: {clt_mean}, CLT Theoretical Std Dev: {clt_std:.4f}")

# Plotting
plt.figure(figsize=(10, 6))
plt.hist(sample_means, bins=30, density=True, alpha=0.6, label='Sampling Distribution of the Mean')

# Overlay theoretical normal curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = stats.norm.pdf(x, clt_mean, clt_std)
plt.plot(x, p, 'k', linewidth=2, label='Normal Curve (from CLT)')

plt.title('Verification of CLT with a Normal Population')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.legend()
plt.show()
````

````python
##19. [cite\_start]Implement hypothesis testing using Z-statistics for a sample dataset.## [cite: 46]


import numpy as np
from statsmodels.stats.weightstats import ztest

# Hypothesis: The mean height of a sample of students is different from the known national average.
# H0: The sample mean is equal to the population mean.
# H1: The sample mean is not equal to the population mean.

# Parameters
population_mean = 170  # Known national average height in cm
alpha = 0.05           # Significance level

# Generate a sample dataset
np.random.seed(42)
sample_data = np.random.normal(loc=172, scale=10, size=50) # Sample with mean 172
sample_mean = np.mean(sample_data)
print(f"Sample Mean: {sample_mean:.2f} cm")

# Perform the Z-test (using statsmodels)
# We assume the population standard deviation is known (e.g., 10 cm)
# In ztest, value is the population mean under H0
z_statistic, p_value = ztest(x1=sample_data, value=population_mean)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"Reject the null hypothesis. There is a significant difference in mean height (p < {alpha}).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean height.")
````

````python
##20. [cite\_start]Create a confidence interval for a dataset using Python and interpret the result.## [cite: 47]


import numpy as np
import scipy.stats as st

# Generate a sample dataset
np.random.seed(10)
data = np.random.normal(loc=80, scale=15, size=40)

# Calculate the 95% confidence interval for the mean
confidence_level = 0.95
degrees_freedom = len(data) - 1
sample_mean = np.mean(data)
sample_standard_error = st.sem(data) # Calculates std(data)/sqrt(n)

# Create the confidence interval using the t-distribution (as population std is unknown)
# For large samples (n>30), t-distribution is very close to normal distribution.
conf_interval = st.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sample_standard_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: {conf_interval}")

# Interpretation
print(f"\nInterpretation: We are 95% confident that the true population mean lies between {conf_interval[0]:.2f} and {conf_interval[1]:.2f}.")
````

````python
##23. [cite\_start]Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.## [cite: 50]


from scipy.stats import poisson
import matplotlib.pyplot as plt
import numpy as np

# Parameter for the Poisson distribution
lambda_param = 5  # Average number of events

# Value for which to calculate the CDF
k = 7 # We want to find P(X <= 7)

# Calculate the CDF
cdf_value = poisson.cdf(k, mu=lambda_param)

print(f"For a Poisson distribution with Î» = {lambda_param}:")
print(f"The probability of observing {k} or fewer events (P(X <= {k})) is: {cdf_value:.4f}")

# Interpretation
print(f"\nInterpretation: There is a {cdf_value*100:.2f}% chance of observing 7 or fewer events in the given interval.")

# Visualize the CDF
k_values = np.arange(0, 15)
cdf_values = poisson.cdf(k_values, mu=lambda_param)

plt.figure(figsize=(8, 5))
plt.step(k_values, cdf_values, where='post')
plt.title(f'CDF of Poisson Distribution (Î»={lambda_param})')
plt.xlabel('Number of Events (k)')
plt.ylabel('Cumulative Probability P(X <= k)')
plt.xticks(k_values)
plt.grid(True)
plt.show()
````

````python
##27. [cite\_start]Implement Z-score normalization on a dataset and visualize its transformation.## [cite: 54]



import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Applies Z-score normalization to a dataset and visualizes the transformation.
    """
    mean = np.mean(data)
    std = np.std(data)
    normalized_data = (data - mean) / std

    # Print stats
    print("--- Original Data ---")
    print(f"Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
    print("\n--- Normalized Data ---")
    print(f"Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")

    # Visualize the transformation
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.hist(data, bins=30, color='skyblue', edgecolor='black')
    ax1.set_title('Original Data Distribution')
    ax1.set_xlabel('Value')
    ax1.set_ylabel('Frequency')

    ax2.hist(normalized_data, bins=30, color='lightgreen', edgecolor='black')
    ax2.set_title('Z-Score Normalized Data')
    ax2.set_xlabel('Z-Score')

    plt.tight_layout()
    plt.show()

# Example Usage
sample_data = np.random.exponential(scale=5.0, size=1000) * 10
z_score_normalization(sample_data)
````

````python
##28. [cite\_start]Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.## [cite: 55]


import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skew_kurtosis(data):
    """
    Calculates and prints the skewness and kurtosis of a dataset.
    """
    data_skewness = skew(data)
    # Fisher's definition of kurtosis is used by default (normal = 0)
    data_kurtosis = kurtosis(data, fisher=True)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (Fisher): {data_kurtosis:.4f}")

    # Interpretation
    if abs(data_skewness) < 0.5:
        print("The distribution is approximately symmetric.")
    elif data_skewness > 0.5:
        print("The distribution is positively skewed (right-tailed).")
    else:
        print("The distribution is negatively skewed (left-tailed).")

    if abs(data_kurtosis) < 0.5:
        print("The kurtosis is mesokurtic (similar to a normal distribution).")
    elif data_kurtosis > 0.5:
        print("The kurtosis is leptokurtic (heavier tails, sharper peak).")
    else:
        print("The kurtosis is platykurtic (lighter tails, flatter peak).")


# Generate data from a normal distribution
# For a perfect normal distribution, skewness is 0 and kurtosis is 0.
# A finite sample will have values close to 0.
np.random.seed(99)
normal_data = np.random.normal(loc=0, scale=1, size=5000)

print("--- Analysis of a Normally Distributed Dataset ---")
calculate_skew_kurtosis(normal_data)

# For comparison, analyze a skewed dataset (Exponential)
print("\n--- Analysis of a Skewed Dataset (Exponential) ---")
skewed_data = np.random.exponential(size=5000)
calculate_skew_kurtosis(skewed_data)
````
