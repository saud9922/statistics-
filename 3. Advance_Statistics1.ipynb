```python
##3. [cite\_start]Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.## [cite: 32]

def bernoulli_pmf(k, p):
    """
    Calculates the Probability Mass Function (PMF) of a Bernoulli distribution.

    Args:
        k (int): The outcome (0 for failure, 1 for success).
        p (float): The probability of success (must be between 0 and 1).

    Returns:
        float: The probability of the outcome k.
    """
    if k not in [0, 1]:
        return 0
    if not (0 <= p <= 1):
        raise ValueError("Probability 'p' must be between 0 and 1.")

    if k == 1:
        return p
    else: # k == 0
        return 1 - p

# Example usage:
p_success = 0.7
prob_of_success = bernoulli_pmf(1, p_success)
prob_of_failure = bernoulli_pmf(0, p_success)

print(f"Probability of success (k=1): {prob_of_success}")
print(f"Probability of failure (k=0): {prob_of_failure}")
```

```python
##9. [cite\_start]Write a Python function to calculate Z-scores from a dataset and plot them.## [cite: 38]


import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a dataset and plots their distribution.

    Args:
        data (array-like): The input dataset.
    """
    mean = np.mean(data)
    std = np.std(data)
    z_scores = (data - mean) / std

    print(f"Mean of original data: {mean:.2f}")
    print(f"Standard Deviation of original data: {std:.2f}")
    print(f"Mean of Z-scores: {np.mean(z_scores):.2f}")
    print(f"Standard Deviation of Z-scores: {np.std(z_scores):.2f}")

    # Plotting
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.hist(data, bins=30, alpha=0.7, color='blue')
    ax1.set_title('Original Data Distribution')
    ax1.set_xlabel('Value')
    ax1.set_ylabel('Frequency')

    ax2.hist(z_scores, bins=30, alpha=0.7, color='green')
    ax2.set_title('Z-Score Distribution')
    ax2.set_xlabel('Z-Score')

    plt.tight_layout()
    plt.show()
    return z_scores

# Example usage with some sample data
sample_data = np.random.normal(50, 10, 1000)
z_scores = calculate_and_plot_z_scores(sample_data)
```

```python
#### **11. Bayesian Inference with Bayes' Theorem**


def bayesian_inference(prior, likelihood, evidence):
    """
    Simple implementation of Bayes' Theorem.
    """
    posterior = (likelihood * prior) / evidence
    return posterior

# Example: Medical Diagnosis
# P(A) = Prior probability of having the disease
# P(B|A) = Probability of testing positive given the disease (sensitivity)
# P(B) = Overall probability of testing positive

p_disease = 0.01
p_positive_given_disease = 0.99 # Sensitivity
p_positive_given_no_disease = 0.05 # False positive rate
p_no_disease = 1 - p_disease

# P(B) = P(B|A)P(A) + P(B|not A)P(not A)
p_positive = (p_positive_given_disease * p_disease) + (p_positive_given_no_disease * p_no_disease)

# Calculate P(Disease | Positive Test)
p_disease_given_positive = bayesian_inference(p_disease, p_positive_given_disease, p_positive)

print(f"Posterior probability of having the disease given a positive test: {p_disease_given_positive:.4f}")
```

```python
#### **12. Chi-square Test for Independence**


import numpy as np
from scipy.stats import chi2_contingency

# Observed data in a contingency table
# Rows: Smoker, Non-smoker; Columns: Lung Cancer, No Lung Cancer
observed = np.array([[50, 10], [20, 120]])

chi2, p, dof, expected = chi2_contingency(observed)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)

if p < 0.05:
    print("\nReject the null hypothesis: There is a significant association between smoking and lung cancer.")
else:
    print("\nFail to reject the null hypothesis.")

```

```python
#### **13. Calculate Expected Frequencies for Chi-square Test**


import numpy as np

def expected_frequencies(observed_table):
    """
    Calculates expected frequencies for a Chi-square test.
    """
    row_totals = observed_table.sum(axis=1)
    col_totals = observed_table.sum(axis=0)
    grand_total = observed_table.sum()
    
    expected_table = np.outer(row_totals, col_totals) / grand_total
    return expected_table

# Example usage
observed = np.array([[30, 10], [15, 25]])
expected = expected_frequencies(observed)
print("Observed Frequencies:\n", observed)
print("\nExpected Frequencies:\n", expected)


```

```python
#### **14. Goodness-of-Fit Test**


from scipy.stats import chisquare

# Observed frequencies of dice rolls
observed = [9, 11, 8, 12, 10, 10] # Total 60 rolls
# Expected frequencies for a fair die
expected = [10, 10, 10, 10, 10, 10]

chi2, p = chisquare(f_obs=observed, f_exp=expected)

print(f"Chi-square statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")

if p < 0.05:
    print("Reject the null hypothesis: The die is likely not fair.")
else:
    print("Fail to reject the null hypothesis: The die appears to be fair.")

```

```python
##19. [cite\_start]Implement hypothesis testing using Z-statistics for a sample dataset.## [cite: 46]


import numpy as np
from statsmodels.stats.weightstats import ztest

# Hypothesis: The mean height of a sample of students is different from the known national average.
# H0: The sample mean is equal to the population mean.
# H1: The sample mean is not equal to the population mean.

# Parameters
population_mean = 170  # Known national average height in cm
alpha = 0.05           # Significance level

# Generate a sample dataset
np.random.seed(42)
sample_data = np.random.normal(loc=172, scale=10, size=50) # Sample with mean 172
sample_mean = np.mean(sample_data)
print(f"Sample Mean: {sample_mean:.2f} cm")

# Perform the Z-test (using statsmodels)
# We assume the population standard deviation is known (e.g., 10 cm)
# In ztest, value is the population mean under H0
z_statistic, p_value = ztest(x1=sample_data, value=population_mean)

print(f"Z-statistic: {z_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Conclusion
if p_value < alpha:
    print(f"Reject the null hypothesis. There is a significant difference in mean height (p < {alpha}).")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in mean height.")
```

```python
##20. [cite\_start]Create a confidence interval for a dataset using Python and interpret the result.## [cite: 47]


import numpy as np
import scipy.stats as st

# Generate a sample dataset
np.random.seed(10)
data = np.random.normal(loc=80, scale=15, size=40)

# Calculate the 95% confidence interval for the mean
confidence_level = 0.95
degrees_freedom = len(data) - 1
sample_mean = np.mean(data)
sample_standard_error = st.sem(data) # Calculates std(data)/sqrt(n)

# Create the confidence interval using the t-distribution (as population std is unknown)
# For large samples (n>30), t-distribution is very close to normal distribution.
conf_interval = st.t.interval(confidence_level, degrees_freedom, loc=sample_mean, scale=sample_standard_error)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval for the Mean: {conf_interval}")

# Interpretation
print(f"\nInterpretation: We are 95% confident that the true population mean lies between {conf_interval[0]:.2f} and {conf_interval[1]:.2f}.")
```

```python
##27. [cite\_start]Implement Z-score normalization on a dataset and visualize its transformation.## [cite: 54]



import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Applies Z-score normalization to a dataset and visualizes the transformation.
    """
    mean = np.mean(data)
    std = np.std(data)
    normalized_data = (data - mean) / std

    # Print stats
    print("--- Original Data ---")
    print(f"Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
    print("\n--- Normalized Data ---")
    print(f"Mean: {np.mean(normalized_data):.2f}, Std Dev: {np.std(normalized_data):.2f}")

    # Visualize the transformation
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.hist(data, bins=30, color='skyblue', edgecolor='black')
    ax1.set_title('Original Data Distribution')
    ax1.set_xlabel('Value')
    ax1.set_ylabel('Frequency')

    ax2.hist(normalized_data, bins=30, color='lightgreen', edgecolor='black')
    ax2.set_title('Z-Score Normalized Data')
    ax2.set_xlabel('Z-Score')

    plt.tight_layout()
    plt.show()

# Example Usage
sample_data = np.random.exponential(scale=5.0, size=1000) * 10
z_score_normalization(sample_data)
```

```python
##28. [cite\_start]Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.## [cite: 55]


import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skew_kurtosis(data):
    """
    Calculates and prints the skewness and kurtosis of a dataset.
    """
    data_skewness = skew(data)
    # Fisher's definition of kurtosis is used by default (normal = 0)
    data_kurtosis = kurtosis(data, fisher=True)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (Fisher): {data_kurtosis:.4f}")

    # Interpretation
    if abs(data_skewness) < 0.5:
        print("The distribution is approximately symmetric.")
    elif data_skewness > 0.5:
        print("The distribution is positively skewed (right-tailed).")
    else:
        print("The distribution is negatively skewed (left-tailed).")

    if abs(data_kurtosis) < 0.5:
        print("The kurtosis is mesokurtic (similar to a normal distribution).")
    elif data_kurtosis > 0.5:
        print("The kurtosis is leptokurtic (heavier tails, sharper peak).")
    else:
        print("The kurtosis is platykurtic (lighter tails, flatter peak).")


# Generate data from a normal distribution
# For a perfect normal distribution, skewness is 0 and kurtosis is 0.
# A finite sample will have values close to 0.
np.random.seed(99)
normal_data = np.random.normal(loc=0, scale=1, size=5000)

print("--- Analysis of a Normally Distributed Dataset ---")
calculate_skew_kurtosis(normal_data)

# For comparison, analyze a skewed dataset (Exponential)
print("\n--- Analysis of a Skewed Dataset (Exponential) ---")
skewed_data = np.random.exponential(size=5000)
calculate_skew_kurtosis(skewed_data)
```
